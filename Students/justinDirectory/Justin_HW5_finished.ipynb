```python
# PART 1.A.

# R = Omega Radiation , V = Omega Lambda , M = Omega Matter 


import matplotlib.pyplot as plt
from scipy.integrate import quad
from scipy.integrate.quadpack import error as quadpack_error
import math 
    
# M = 2.011 , V = R = 0 is big crunch universe goes up to a = 1.98


# my output is time and input is a(t) (this is really weird), so I switch them in the plt.plot at the very end but -
# I do all my analysis by thinking of plots of time vs a(t)

# My strategy here is to analyze how t changes with values of 'a(t)' in the integral. Hence I create a variable del_t which -
# functions as a quasi derivative. If I see that as 'a(t) increases del_t with respect to 'a(t)' goes from small to large and - 
# back to small then I must have a loitering universe. The green dot represents the time at which a(t) is constant


max_cat = []
H_0 = 2.2037304 * 10 ** - 18
test = 'no'
for V in range(0 , 10):
    V = V/10
    for M in range(1 , 10):
        M = M/10
        for R in range(0 , 10):
            R = R/10
            def timefunc(a):
                f = 1/(H_0 * (R/(a ** 2) + M/a + V * a**2 + 1 - V - M) ** 0.5)
                return f
            behave = []
            input_cat = []
            output_cat = []
            prev_time = 0
            divs = 1000
            #print(V , M , R)
            for upper_limit in range(1 ,5 * divs):
                uplimdiv = upper_limit / divs
                time , error = quad(timefunc , 0 , uplimdiv)
                time_gyr = time * 3.17098e-17
                #print('a = ' , uplimdiv , '||' , 'time = ', time_gyr , 'Gyr')
                del_t = time_gyr - prev_time
                behave.append(del_t)
                #print(del_t)
                prev_time = time_gyr
                output_cat.append(time_gyr)
                input_cat.append(uplimdiv)
            check = 'no'
            if behave[2] > behave[1]:
                for n in range(2 , len(behave)):
                    if behave[n - 1] > behave[n]:
                        check = 'yes'
                        max_cat.append(behave[n - 1])
                        a_input = (n - 1)/divs
                        #print(behave[n - 1] , a_input , 'max')
                        fit = math.isclose(a_input , 1.0 , rel_tol = .05 , abs_tol = 0.0)
                        if fit == True :
                            fig = plt.figure()
                            fig.suptitle('growth of universe' + '  M = ' + str(M) + '  V = ' + str(V) + '  R = ' + str(R), fontsize = 20)
                            plt.ylabel('value of a(t)')
                            plt.xlabel('time in Gyr')
                            plt.xlim(0 , 50)
                            plt.ylim(0 , 5)
                            plt.plot(output_cat , input_cat)
                            plt.plot(output_cat[n - 1] , input_cat[n - 1] , marker = 'o')
                            plt.show()
                            test = 'yes'
                            break
                if test == 'yes':
                    break
            if test == 'yes':
                break
        if test == 'yes':
            break
    if test == 'yes':
        break

#comment all 'if test' statements for multiple loitering universes; warning their are many many loitering universes
```
