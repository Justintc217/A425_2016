p = density = 10$^{-9}$ Mpc$^{-3}$

r = radius of star = 7 * 10 ^ 8

R = radius from earth to shell

p = N/((4/3)(pi)R^3)
N = number of stars

Start by creating an infinite set of spherical shells of varying R and thickness dR; these can fill the entire universe

Rearange density equation and take derivative 

dN = p(4(pi)R^2)dR $ (1.0)

$a=\frac{b}{c}$

This is the number density of stars per shell

quantity 1: p(pi)r^2 dN 

is amount of arbitrary shell that is blocked. pdN tells us the amount of stars and (pi)r^2 is how much area each star blocks

quantity 2: 4(pi)R^2 dR 

is the amount of an arbitrary shell that is open.

dividing by the two to find the ratio of blocked amount per shell we divide quantity 1 by quantity 2 and by using dN substitution (eq 1.0) gives us:

p(pi)r^2 

integrating this "percentage of blocking per shell" through all shells from R = 0 to R = infinity gives us:

Rp(pi)r^2 

now to find average distance we divide this by the radius of all shells again from R = 0 to R = infinity which gives us:

(Rp(pi)r^2)/R = p(pi)r^2

when we plug in the above numerical values we find

p(pi)r^2 = 5.2379 * 10^-41 meter^-1

remembering to change Mpc^3 to m^3

PROBLEM 3

using: ds^2 = dr^2 + R^2 * sin^2(r/R)[dθ^2 + sin^2(θ)dφ^2]

back solve for dθ. dφ = 0 since we are only interested in the width of the object

algebra...

dθ^2 = (ds^2 - dr^2)/(R^2 * sin^2(r/R))

square root

dθ = [sqrt(ds^2 - dr^2)]/(Rsin(r/R))

as r goes to (pi)R the denominator goes to 0 since sin(pi) = 0

This means dθ goes to infinity.

This makes sense since as an arbitrary object gets closer to the center of the polar coordinates then slight changes in r can result in huge changes in θ. (eg. crossing the pole can flip θ by 180 degrees). Thus dθ would go to infinity when the object is at a pole

```python
# PROBLEM 4 CONTINUED
how_far = 1
import matplotlib.pyplot as plt
for show_4 in range (0 , 4):
    t_values = []
    how_far = how_far * 10
    for x in range(0, how_far):
        t_values.append(x)


    a = [(3.79196 * t **(0.4)) for t in t_values]

    plt.plot(t_values, a)

    plt.draw()

    plt.show()
```

```python
#PROBLEM 4: SOLVING NUMERICALLY

# a' = 70a^-(1.5)
#integrate (a^1.5) da = 70t
# we can append t at the end so we really are numerically solving integral from 0 to x a^1.5 = 70


import matplotlib.pyplot as plt
from scipy.integrate import quad

def integrand(a):
    return a**(1.5)

x = 1
t_cat = []
a_cat = []
while x < (100):
    x = x + 1

    A = x / 10.0

    time, err = quad(integrand, 0, A)
    time = time/70.0
    t_cat.append(time)
    a_cat.append(A)






plt.plot(t_cat, a_cat)

plt.draw()

plt.show()
    
```
